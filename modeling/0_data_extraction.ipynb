```python
import kagglehub
import os
import shutil

path = kagglehub.competition_download('home-credit-default-risk')
print("Path to competition files:", path)
dest_dir = 'input_data/row_input_data'

os.makedirs(dest_dir, exist_ok=True)

for filename in os.listdir(path):
    src_file = os.path.join(path, filename)
    dest_file = os.path.join(dest_dir, filename)
    
    if os.path.isfile(src_file):
        shutil.copy2(src_file, dest_file)
        print(f"Copied: {filename}")
    elif os.path.isdir(src_file):
        if os.path.exists(dest_file):
            shutil.rmtree(dest_file)
        shutil.copytree(src_file, dest_file)
        print(f"Copied directory: {filename}")

print(f"\nAll files extracted to: {os.path.abspath(dest_dir)}")
```

```python
dest_dir = 'input_data/row_input_data'
print("Files in destination directory:")
print("=" * 50)

for root, dirs, files in os.walk(dest_dir):
    level = root.replace(dest_dir, '').count(os.sep)
    indent = ' ' * 2 * level
    print(f"{indent}{os.path.basename(root)}/")
    
    subindent = ' ' * 2 * (level + 1)
    for file in files:
        file_path = os.path.join(root, file)
        file_size = os.path.getsize(file_path) / (1024 * 1024)  
        print(f"{subindent}{file} ({file_size:.2f} MB)")

print("\n" + "=" * 50)
print(f"Total files extracted: {sum(len(files) for _, _, files in os.walk(dest_dir))}")
```
