## Uploading Feature Selected Data - Neon Database

```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
load_dotenv()

ROOT_PATH = "input_data/row_fs_data"
input_file_names = [file for file in os.listdir(ROOT_PATH) if file.endswith(".csv")]
DATABASE_URL = os.getenv("DATABASE_URL")

if not DATABASE_URL:
    raise ValueError(
        "DATABASE_URL missing! Please check your .env file."
    )
engine = create_engine(DATABASE_URL)

for file_name in input_file_names:
    print(f"uploadeding {file_name}...")
    file_path = os.path.join(ROOT_PATH, file_name)
    df = pd.read_csv(file_path)
    table_name = os.path.splitext(file_name)[0]
    df.to_sql(
        table_name,
        engine,
        if_exists="replace",
        index=False,
        chunksize=5000,
        method="multi",
    )

    print(f"Table '{table_name}' uploaded successfully from {file_name}...")
```
