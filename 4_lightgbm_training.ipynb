```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import os
import re
```

```python
ROOT_PATH = "input_data/row_fs_data"
```

```python
app_train = pd.read_csv(os.path.join(ROOT_PATH, "app_train_fs.csv"))
app_test = pd.read_csv(os.path.join(ROOT_PATH, "app_test_fs.csv"))

print(f"Shape of app_train : {app_train.shape}")
print(f"Shape of app_test  :  {app_test.shape}")
```

```python
def get_date_columns(columns):
    """
    Identify date-related columns based on column names.
    """

    date_pattern = re.compile(
        r"(date|dt|dob|birth|created|updated|modified|opened|closed|expiry|expiration|issued|renewal|start|timestamp|time)",
        re.IGNORECASE
    )

    return [col for col in columns if date_pattern.search(col)]

date_cls = get_date_columns(app_train.columns)
print(f"Total Time Columns: ", len(date_cls))
app_train[date_cls].sample(5)
```
